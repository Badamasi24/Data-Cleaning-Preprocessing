```python
import pandas as pd
df=pd.read_csv('2) Stock Prices Data Set.csv')
```

```python
df.info()
```

```python
df.head()
```

**Handling missing values**

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
missing_cols=['open','high','low']
for col in missing_cols:
    df[col]=df[col].fillna(df[col].mean())
```

```python
df.isnull().any()
```

**Handling duplicate value**

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

**changing date column to date/time**

```python
df['date']
```

```python
df['date']=pd.to_datetime(df['date'])
```

**convert float to int type
**

```python
float_col =['open','high','low','close','volume']
for col in float_col:
  df[col]=df[col].round().astype(int)
```
